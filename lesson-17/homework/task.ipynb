***2x2 grid of subplots***

```python
# Top-left: $ f(x) = x^3 $
#Top-right: $ f(x) = \sin(x) $
#Bottom-left: $ f(x) = e^x $
#Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)
import numpy as np
import matplotlib.pyplot as plt

# Define the x values for different functions
x1 = np.linspace(-2, 2, 100)  # For x^3
x2 = np.linspace(-2 * np.pi, 2 * np.pi, 100)  # For sin(x)
x3 = np.linspace(0, 2, 100)  # For e^x
x4 = np.linspace(0, 10, 100)  # For log(x+1)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#  f(x) = x^3
axs[0, 0].plot(x1, x1**3, 'r-', label='$f(x) = x^3$')
axs[0, 0].set_title("Cubic Function")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("$f(x)$")
axs[0, 0].grid(True)
axs[0, 0].legend()

#  f(x) = sin(x)
axs[0, 1].plot(x2, np.sin(x2), 'g-', label='$f(x) = \sin(x)$')
axs[0, 1].set_title("Sine Function")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("$f(x)$")
axs[0, 1].grid(True)
axs[0, 1].legend()

#  f(x) = e^x
axs[1, 0].plot(x3, np.exp(x3), 'b-', label='$f(x) = e^x$')
axs[1, 0].set_title("Exponential Function")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("$f(x)$")
axs[1, 0].grid(True)
axs[1, 0].legend()

#  f(x) = log(x+1)
axs[1, 1].plot(x4, np.log(x4 + 1), 'm-', label='$f(x) = \log(x+1)$')
axs[1, 1].set_title("Logarithmic Function")
axs[1, 1].set_xlabel("x")
axs[1, 1].set_ylabel("$f(x)$")
axs[1, 1].grid(True)
axs[1, 1].legend()


plt.tight_layout()
plt.show()
```

 Scatter Plot

```python
#Task: Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid.

import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0, 10, 100)  
y = np.random.uniform(0, 10, 100) 
sizes = np.random.uniform(20, 200, 100)  
colors = np.random.uniform(0, 10, 100)  
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7, edgecolors='k')

plt.colorbar(label='Color scale')

plt.title('Scatter Plot of 100 Random Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.show()
```
